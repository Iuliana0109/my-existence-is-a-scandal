```python
import sys
import time

def time_ms():
    return time.time() * 1000

class Graph():

    def __init__(self, vertices):
        self.V = vertices
        self.graph = [[0 for column in range(vertices)] 
                    for row in range(vertices)]

    def addEdge(self, u, v, w):
        self.graph[u][v] = w
        self.graph[v][u] = w

    def minKey(self, key, mst_set):

        # Initilaize min value
        INF = sys.maxsize

        for v in range(self.V):
            
            if key[v] < INF and mst_set[v] == False:
                INF = key[v]
                min_index = v

        return min_index

    def Prim(self):

        key = [sys.maxsize] * self.V
        mst = [0] * self.V  # Array to store constructed MST

        key[0] = 0
        mst_set = [False] * self.V

        mst[0] = -1  # First node - the root node

        for _ in range(self.V):

            u = self.minKey(key, mst_set)

            mst_set[u] = True
            
            for v in range(self.V):

                if self.graph[u][v] > 0 and mst_set[v] == False and key[v] > self.graph[u][v]:
                    key[v] = self.graph[u][v]
                    mst[v] = u

        self.printMST(mst)
        
    def printMST(self, mst):
            
        print("Edges   Weight")
        
        min_weigth = 0
        
        for i in range(1, self.V):
            min_weigth += self.graph[i][mst[i]]
            print(mst[i], "--", i, "==", self.graph[i][mst[i]])
            
        print('Total weight:', min_weigth)
```

```python
import time

def time_ms():
    return time.time() * 1000

class Graph:
    def __init__(self, vertices):
        self.V = vertices  # No. of vertices for each edge
        self.graph = []         
        
    def addEdge(self, u, v, w): 
        self.graph.append([u, v, w])  
          
    def search(self, parent, i):
        if parent[i] == i: return i
        return self.search(parent, parent[i])
    
    def union(self, parent, lvl, x, y): #union function
        x_root = self.search(parent, x)
        y_root = self.search(parent, y)
        if lvl[x_root] < lvl[y_root]: parent[x_root] = y_root         
        elif lvl[x_root] > lvl[y_root]: parent[y_root] = x_root
        else:
            parent[y_root] = x_root
            lvl[x_root] += 1
            
    def Kruskal(self): # The main function to construct MST using Kruskal's Algorithm
        mst = [] 
        i, idx = 0,0
        self.graph = sorted(self.graph, key=lambda item: item[2]) 
        parent = []
        lvl = []
        
        for node in range(self.V):
            parent.append(node)
            lvl.append(0)      
            
        while idx < self.V - 1:         
            u, v, w = self.graph[i]
            i = i + 1
            x = self.search(parent, u)
            y = self.search(parent, v)           
            if x != y: #if u and v are not of the same tree, apply union
                idx = idx + 1
                mst.append([u, v, w])
                self.union(parent, lvl, x, y)  
                 
        self.printMST(mst)
        
    def printMST(self,mst):   
             
        print("Edges   Weight")
        
        min_weight = 0
        
        for u, v, weight in mst:
            min_weight += weight
            print(u, '--', v, '==', weight)
            
        print("Total weight:", min_weight)
```
