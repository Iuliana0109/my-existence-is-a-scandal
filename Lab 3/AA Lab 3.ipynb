#LAB3: Eratosthenes Sieve

```python
#1: Sieve of Eratosthenes
def sieve(n):
    # Create a list of booleans initialized to True
    c = [True] * (n+1)
    c[0] = False
    c[1] = False
    
    i = 2
    while (i <= n):
        if (c[i] == True):
            j = 2 * i
            while (j <= n):
                c[j] = False
                j = j + i
        i = i + 1
    
    # Return a list of primes
    primes = []
    for i in range(2, n+1):
        if c[i] == True:
            primes.append(i)
    
    return primes
```

```python
#2: Optimized Sieve of Erathostenes

def optimized_sieve_of_eratosthenes(n):
    c = [True] * (n + 1)
    c[0] = c[1] = False
    i = 2
    while i * i <= n:
        j = i * 2
        while j <= n:
            c[j] = False
            j += i
        i += 1
    return [i for i in range(n+1) if c[i]]
```

```python
#3: Sieve of Sundaram

def sieve_of_sundaram(n):
    k = (n - 1) // 2
    c = [False] * (k + 1)
    for i in range(1, k+1):
        j = i
        while i+j+2*i*j <= k:
            c[i+j+2*i*j] = True
            j += 1
    if n > 2:
        return [2] + [2*i+1 for i in range(1, k+1) if not c[i]]
    else:
        return [2]
```

```python
#4: Brute Force Method

def brute_force_method(n):
    primes = []
    for i in range(2, n+1):
        is_prime = True
        for j in range(2, i):
            if i % j == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(i)
    return primes
```

```python
#5: Trial Division Method

def trial_division_method(n):
    primes = []
    for i in range(2, n+1):
        is_prime = True
        for j in range(2, int(i**0.5)+1):
            if i % j == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(i)
    return primes
```
