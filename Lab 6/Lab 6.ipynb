```python
import time
from decimal import Decimal
from decimal import getcontext

def time_ms():
    return time.time() * 1000

def pi_d(n):
    
    pi = Decimal(0)
    
    getcontext().prec=n+1
    
    pi = sum(1/Decimal(16)**k * 
        (Decimal(4)/(8*k+1) - Decimal(2)/(8*k+4) - 
         Decimal(1)/(8*k+5) - Decimal(1)/(8*k+6)) for k in range (n))
  
    return pi

n=240
start_BBP = time_ms()    
result=str(pi_d(n))
print('i-th digit: ',result[-1])
end_BBP = time_ms()
print("\nThe time of execution for BBP:", end_BBP - start_BBP)
```

```python
import time
from math import factorial
from decimal import Decimal, getcontext


def time_ms():
    return time.time() * 1000

def pi_d(n):
    
    S = Decimal(0)
    pi = Decimal(0)
    denominator = Decimal(0)
    getcontext().prec=n+1
    
    for k in range(n):
        
        S = ((-1)**k)*(factorial(6*k))*(545140134*k+13591409)
        denominator = factorial(3*k)*(factorial(k)**3)*(640320**(3*k))
        pi += Decimal(S)/Decimal(denominator)
        
    pi = pi * Decimal(12) / Decimal(640320 ** Decimal(3/2))
    pi = 1/pi
    
    return pi

n = 10
start_Chudnovsky = time_ms() 
result = str(pi_d(n))
print('i-th digit: ',result[-1])
end_Chudnovsky = time_ms()
print("\nThe time of execution for Chudnovsky:", end_Chudnovsky - start_Chudnovsky)
```
